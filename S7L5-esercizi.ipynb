```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

```

```python
#carica il dataset
df = pd.read_csv('C:\\Users\\Isabe\\OneDrive\\Documentos\\PYTHON\\CursoPython\\dataset_climatico.csv')
```

```python
# PULIZIA DEI DATI: Rimuovi righe con valori mancanti
df.dropna(inplace=True)
```

```python
# calcolare la media e la dev standard
media_temperatura = df['temperatura_media'].mean()
std_temperatura = df['temperatura_media'].std()

media_precipitazioni = df['precipitazioni'].mean()
std_precipitazioni = df['precipitazioni'].std()

media_umidita = df['umidita'].mean()
std_umidita = df['umidita'].std()

media_velocita_vento = df['velocita_vento'].mean()
std_velocita_vento = df['velocita_vento'].std()
```

```python
#Applicare la normalizzazione Z-score
df['temperatura_media_n'] = (df['temperatura_media'] - media_temperatura) / std_temperatura
df['precipitazioni_n'] = (df['precipitazioni'] - media_precipitazioni) / std_precipitazioni
df['umidita_n'] = (df['umidita'] - media_umidita) / std_umidita
df['velocita_vento_n'] = (df['velocita_vento'] - media_velocita_vento) / std_velocita_vento
```

```python
#Salvare dataset normalizzato
df.to_csv('C:\\Users\\Isabe\\OneDrive\\Documentos\\PYTHON\\CursoPython\\dataset_climatico_normalizzato.csv', index=False) 
```

```python
# ANALISI ESPLORATIVA DEI DATI
statistiche = df[['temperatura_media', 'precipitazioni', 'umidita', 'velocita_vento']].describe()
```

```python
# Istogrammi delle variabili
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.histplot(df['temperatura_media'], kde=True)
plt.title('Istogramma - Temperatura Media')
plt.show()
```

```python
plt.subplot(2, 2, 2)
sns.histplot(df['precipitazioni'], kde=True)
plt.title('Istogramma - Precipitazioni')
plt.show()
```

```python
plt.subplot(2, 2, 3)
sns.histplot(df['umidita'], kde=True)
plt.title('Istogramma - Umidit')
plt.show()
```

```python
plt.subplot(2, 2, 4)
sns.histplot(df['velocita_vento'], kde=True)
plt.title('Istogramma - Velocit Vento')

plt.tight_layout()
plt.show()
```

```python
# Box plots delle variabili
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.boxplot(y=df['temperatura_media'])
plt.title('Box Plot - Temperatura Media')

plt.subplot(2, 2, 2)
sns.boxplot(y=df['precipitazioni'])
plt.title('Box Plot - Precipitazioni')

plt.subplot(2, 2, 3)
sns.boxplot(y=df['umidita'])
plt.title('Box Plot - Umidit')

plt.subplot(2, 2, 4)
sns.boxplot(y=df['velocita_vento'])
plt.title('Box Plot - Velocit Vento')

plt.tight_layout()
plt.show()
```

```python
#ANALISI DI CORRELAZIONE

correlazione = df[['temperatura_media', 'precipitazioni', 'umidita', 'velocita_vento']].corr()
```

```python
#Creare la heatmap

plt.figure(figsize=(8, 6))
sns.heatmap(correlazione, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap di Correlazione tra Variabili Meteorologiche')
plt.show()
```
